## Handling Missing Values

```python
df = data
```

```python
# df = df.dropna(subset=['Married']) (drops rows with NULL values in 'Married' column
df['Married'].mode()
```

```python
df['Gender'].mode()
```

```python
df['Dependents'].mode()
```

```python
df['Self_Employed'].mode()
```

```python
df['Loan_Amount_Term'].mode()
```

```python
df['Credit_History'].mode()
```

```python
df[['Married', 'Gender', 'Dependents', 'Self_Employed', 'Loan_Amount_Term', 'Credit_History']] = \
df[['Married', 'Gender', 'Dependents', 'Self_Employed', 'Loan_Amount_Term', 'Credit_History']].fillna(
    df[['Married', 'Gender', 'Dependents', 'Self_Employed', 'Loan_Amount_Term', 'Credit_History']].mode().iloc[0]
)
```

```python
df['LoanAmount'].mean()
```

```python
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].mean())
```

```python
df.info()
```

```python
df.isnull().sum()
```

### All the null values have been replaced

### Convert float to int data type for better efficiency of code

```python
df['CoapplicantIncome'] = df['CoapplicantIncome'].astype(int)
df['LoanAmount'] = df['LoanAmount'].astype(int)
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].astype(int)
df['Credit_History']=df['Credit_History'].astype(int)
```
