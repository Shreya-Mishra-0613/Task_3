```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("raw_data\crop_yield.csv")
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()
numeric_df = df.select_dtypes(include=['float64', 'int64'])
scaled_df = scaler.fit_transform(numeric_df)
df_scaled= pd.DataFrame(scaled_df, columns=numeric_df.columns)
```

```python
df.head()
```

```python
df_scaled.head()
```

```python
df_scaled.describe()
```

## Effect of scaling

```python
# sns.kdeplot(df['Area'], label="Area", color="brown")
sns.kdeplot(df['Production'], label="Production", color="blue")
sns.kdeplot(df['Yield'], label="Yield", color="green")
plt.legend(title="Datasets")
plt.title('Before Scaling')
```

```python
# sns.kdeplot(df_scaled['Area'], label="Area", color="brown")
sns.kdeplot(df_scaled['Production'], label="Production", color="blue")
sns.kdeplot(df_scaled['Yield'], label="Yield", color="green")
plt.legend(title="Datasets")
plt.title('After Scaling')
```

```python
sns.kdeplot(df['Fertilizer'], label="Fertilizers used", color="red")
sns.kdeplot(df['Pesticide'], label="Pestisides used", color="green")
plt.legend(title="Datasets")
plt.title('Before Scaling')
plt.show()
sns.kdeplot(df_scaled['Fertilizer'], label="Fertilizers used", color="red")
sns.kdeplot(df_scaled['Pesticide'], label="Pestisides used", color="green")
plt.legend(title="Datasets")
plt.title('After Scaling')
```

```python
sns.kdeplot(df_scaled['Fertilizer'], label="Fertilizers used", color="red")
sns.kdeplot(df_scaled['Area'], label="area", color="green")
plt.legend(title="Datasets")
plt.title('After Scaling')
```
